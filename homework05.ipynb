```python
def GaussieNoisy(image,sigma):
    row,col,ch=image.shape
    mean=0
    gauss=np.random.normal(mean,sigma,(row,col,ch))
    gauss=gauss.reshape(row,col,ch)
    noisy=image+gauss
    return noisy.astype(np.uint8)
```

```python
from tkinter import *
from skimage import io
import numpy as np
from PIL import Image


im=io.imread(r'D:\5_1.jpg', as_grey=True)
im_copy_med = io.imread(r'D:\5_1.jpg', as_grey=True)#几何
im_copy_mea = io.imread(r'D:\5_1.jpg', as_grey=True)#算数
im_copy_met = io.imread(r'D:\5_1.jpg', as_grey=True)#谐波
im_copy_mes = io.imread(r'D:\5_1.jpg', as_grey=True)#逆谐波
#io.imshow(im)
for i in range(0,im.shape[0]):
    for j in range(0,im.shape[1]):
        im_copy_med[i][j]=im[i][j]
        im_copy_mea[i][j]=im[i][j]
        im_copy_met[i][j]=im[i][j]
        im_copy_mes[i][j]=im[i][j]
#ui
root = Tk()
root.title("rdz's homework")
root.geometry('300x200')

medL = Label(root, text="几何均值滤波：")
medL.pack()
med_text = StringVar()
med = Entry(root, textvariable = med_text)
med_text.set("")
med.pack()

meaL = Label(root, text="算数均值滤波：")
meaL.pack()
mea_text = StringVar()
mea = Entry(root, textvariable = mea_text)
mea_text.set("")
mea.pack()

metL = Label(root, text="谐波均值滤波：")
metL.pack()
met_text = StringVar()
met = Entry(root, textvariable = met_text)
met_text.set("")
met.pack()

mesL = Label(root, text="逆谐波均值滤波：")
mesL.pack()
mes_text = StringVar()
mes = Entry(root, textvariable = mes_text)
mes_text.set("")
mes.pack()

def m_filter(x, y, step):#几何均值滤波
    sum_d=1.0
    for k in range(-int(step/2),int(step/2)+1):
        for m in range(-int(step/2),int(step/2)+1):
            sum_d*=(im[x+k][y+m])
    return (sum_d**(1./(step*step)))

def mean_filter(x, y, step):#算数均值滤波
    sum_a = 0
    for k in range(-int(step/2),int(step/2)+1):
        for m in range(-int(step/2),int(step/2)+1):
            sum_a += im[x+k][y+m]
    return sum_a/(step*step)

def met_filter(x, y, step):#谐波均值滤波
    sum_t = 0
    for k in range(-int(step/2),int(step/2)+1):
        for m in range(-int(step/2),int(step/2)+1):
            sum_t *=(1./im[x+k][y+m])
    return ((step*step)/sum_t)

def mes_filter(x, y, step):#逆谐波均值滤波
    sum_s = 1.0
    Q=0.5
    for k in range(-int(step/2),int(step/2)+1):
        for m in range(-int(step/2),int(step/2)+1):
            sum_s *= im[x+k][y+m]
    return ((sum_s**(Q+1))/(sum_s**Q))

def on_click():
    if(med_text):
        medStep = int(med_text.get())
        for i in range(int(medStep/2),im.shape[0]-int(medStep/2)):
            for j in range(int(medStep/2),im.shape[1]-int(medStep/2)):
                im_copy_med[i][j] = m_filter(i, j, medStep)
    if(mea_text):
        meaStep = int(mea_text.get())
        for i in range(int(meaStep/2),im.shape[0]-int(meaStep/2)):
            for j in range(int(meaStep/2),im.shape[1]-int(meaStep/2)):
                im_copy_mea[i][j] = mean_filter(i, j, meaStep)
    if(met_text):
        metStep = int(met_text.get())
        for i in range(int(metStep/2),im.shape[0]-int(metStep/2)):
            for j in range(int(metStep/2),im.shape[1]-int(metStep/2)):
                im_copy_met[i][j] = met_filter(i, j, metStep)
    if(mes_text):
        mesStep = int(mes_text.get())
        for i in range(int(mesStep/2),im.shape[0]-int(mesStep/2)):
            for j in range(int(mesStep/2),im.shape[1]-int(mesStep/2)):
                im_copy_mes[i][j] = mes_filter(i, j, mesStep)
    subplot(221)
    io.imshow(im_copy_med)
    io.imsave(str(medStep) + 'med.jpg', im_copy_med)
    subplot(222)
    io.imshow(im_copy_mea)
    io.imsave(str(meaStep) + 'mea.jpg', im_copy_mea)
    subplot(223)
    io.imshow(im_copy_met)
    io.imsave(str(metStep) + 'met.jpg', im_copy_met)
    subplot(224)
    io.imshow(im_copy_mes)
    io.imsave(str(mesStep) + 'mes.jpg', im_copy_mes)

Button(root, text="Go", command = on_click).pack()

root.mainloop()
```
